3.세공수준별 색상별 가격과 케럿의 평균을 구하고 1캐럿당 가격이 
가장 높은 세공수준과 색상 조합을 순서대로 고르면?

```python
dia = pd.read_csv("실기 Python 실습파일 (2)/diamonds.csv")
dia.head(2)
```

```python
dia_agg = dia.groupby(["cut","color"])[["price","carat"]].mean().reset_index()
dia_agg.head(2)
```

```python
dia_agg["ratio"] = dia_agg["price"] / dia_agg["carat"]
dia_agg = dia_agg.sort_values("ratio", ascending= False)
dia_agg.head()
```
