```python
import pandas as pd
from scipy.stats import pearsonr
from scipy.stats import spearmanr
from scipy.stats import kendalltau
```

```python
df.corr()
```

```python
df[["casual","registered","count"]].corr()
```

```python
df[["casual","registered","count"]].corr(method = "spearman")
```
