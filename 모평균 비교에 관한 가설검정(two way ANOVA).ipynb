```python
import pandas as pd
from statsmodels.formula.api import ols
from statsmodels.stats.anova import anova_lm
```

```python
df = pd.read_csv("실기 Python 실습파일 (2)/diamonds.csv")
df.head(2)
```

```python
formula = "price ~ cut + color +cut:color" #cut:color 교호작용
model = ols(formula, data = df).fit()
anova_lm(model)
```

```python
df = pd.read_csv("실기 Python 실습파일 (2)/bike.csv")
df.head(2)
```

```python
formula = "registered ~ season+holiday + season:holiday" 
model = ols(formula, data = df).fit()
anova_lm(model)
```

```python
formula = "registered ~ C(season)+C(holiday) + C(season):C(holiday)" 
model = ols(formula, data = df).fit()
anova_lm(model)
```

Q2.

```python
formula = "registered ~ C(season)+C(weather) + C(season):C(weather)" 
lm = ols(formula, data = df).fit()
anova_lm(lm)
```

```python
anova_lm(lm).round(5)
```

```python
dia = pd.read_csv("실기 Python 실습파일 (2)/diabetes.csv")
dia.head(2)
```

```python
dia = dia.loc[(dia["Age"]<70)&(dia["BMI"]>0)]
```

```python
dia["is_preg"] = (dia["Pregnancies"] > 0) + 0
dia["Age_g"] = (dia["Age"] // 10) *10
dia.head(2)
```

```python
formula = "BMI ~ C(Age_g)+C(is_preg) + C(Age_g):C(is_preg)" 
lm = ols(formula, dia).fit()
anova_lm(lm).round(5)
```
